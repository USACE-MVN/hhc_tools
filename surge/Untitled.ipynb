```python
def clip(data,box):
    clipped_data=[]
    for i in data:
        if (i[0]>box[0]) and (i[0]<box[1]) and (i[1]<box[2]) and (i[1]>box[3]):
            clipped_data.append(i)
    return clipped_data
```

```python
box=[-94.1,-87.0,32,28.4]
```

```python
sorted_test_clip=clip(sorted_test,box)
```
